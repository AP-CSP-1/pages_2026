
### Checkpoint Review Ticket 

Show your teacher that you have chronicled your accomplishments for each checkpoint.

- Sixty. Each individual should spend 60% of time toward team goals
- Forty. Each individual should spend 40% of timme toward individual mastery.

#### Team Checkpoints

Teams should show progress each week in a live checkpoint review.  It is encourgaged to seek an early opportunity to checkpoint team progress.

- Demos are a great way to show team progress.
  - Design demos can be shown with Design documentation or Jupyter Notebooks
  - FE demos should be shown using Inspect, prior to and after integration.
  - BE demos should be shown using Postman, prior to and after integration.
  - Full Stack demos are always nice, but are not alway required to show incremental progress.
- Burndown list are a great way to show progress within and Issue.
- Kanban boards are a great way to show progress of groups of Issues.


#### Individual Mastery

During checkpoints student should relate their work to individual mastery.  

1. **Code Mastery**
   - Code reviews are a great way to show individual progress.
   - Individual **key** commits are a great way to highlight accomplishments.
   - Analytics are a complement to show code participation.
   - Show your ability to work in the programming languages you have learned.
   - Include code samples or project snippets that highlight your skills.

2. **Team Teaching Knowledge**
  - Paticipating in teach lessons will be part of your individual mastery.
  - Talking about code in in Teaching highlight your individual knowledge.
  - Participating in grading and giving quality feedback is an excellent way to demonstrate knowledge.
  - Explain how your Team project aligns with the College Board requirements shows depth of understanding.
  - Providing specific examples from your project in lesson shows your ability to analyze AP standards.

3. **Tools and Framework Mastery**:
   - Demonstrate your proficiency with debugging and fixing issues show progress of becoming a developer. 
   - Provide examples of how you have utilized various panels in the Browser inspect demoonstrates knowlege.
   - Building test cases in Postman to validate backend APIs shows growth in understanding SDLC.
   - Showing ability to research or incorporate third party tools shows mastery.
   - Building backend UI's for Administrative UIs, backing up, and restoring data shows depth of understanding.
